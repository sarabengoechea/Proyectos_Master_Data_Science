JSON being independent of a platform is one of the major choices of format for transfer of data between computers that exchange tons of data each day. It is easy to read an fast for computer process and is processed with JavaScript parser

## YAML

```python
# En este caso, debemos definir nosotros una función que lea dicho formato, que sería la 
# siguiente:

import yaml
def read_yaml_file(filename):
    with open(filename, 'r') as stream:
        try:
            print(yaml.safe_load(stream))
        except yaml.YAMLError as exc:
            print(exc)
```

```python
users = [{'name': 'John Aherne', 'occupation': 'Marketing Manager'},
         {'name': 'Jonathan Matzinger', 'occupation': 'Student'},
        { 'name': 'Raphael Marceau', 'occupation': 'Auditor'}]
```

```python
with open('users.yaml', 'w') as f:
    
    data = yaml.dump(users, f)
```
